# Student Performance Analysis

```python
import pandas as pd
```

```python
df=pd.read_csv("student_performance.csv")
```

## Data Preparation

```python

df['total_score'] = df['math_score'] + df['programming_score']
print("Added 'total_score' column.")


print("\nDataFrame Info:")
df.info()
```

```python

total_students = len(df)
print(f"Total number of students: {total_students}")


total_columns = df.shape[1]
print(f"Total number of columns: {total_columns}")


unique_departments_count = df['department'].nunique()
print(f"Number of unique departments: {unique_departments_count}")
print("Unique Departments:", df['department'].unique().tolist())
```

```python

department_counts = df['department'].value_counts().reset_index()
department_counts.columns = ['Department', 'Number of Students']
print("Count of students in each department:")
display(department_counts)
```

```python

highest_math_score = df['math_score'].max()
print(f"Highest Math Score: {highest_math_score}")

lowest_math_score = df['math_score'].min()
print(f"Lowest Math Score: {lowest_math_score}")


avg_math_score = df['math_score'].mean()
print(f"Average Math Score: {avg_math_score:.2f}")


avg_programming_score = df['programming_score'].mean()
print(f"Average Programming Score: {avg_programming_score:.2f}")


avg_attendance_percentage = df['attendance_percentage'].mean()
print(f"Average Attendance Percentage: {avg_attendance_percentage:.2f}%")
```

```python

avg_total_score_by_gender = df.groupby('gender')['total_score'].mean().reset_index()
avg_total_score_by_gender.columns = ['Gender', 'Average Total Score']
print("Average total score by gender:")
display(avg_total_score_by_gender)
```

```python

top_5_performers = df.sort_values(by='total_score', ascending=False).head(5)
print("Top 5 students based on total score:")
display(top_5_performers[['name', 'department', 'total_score']])
```

## Attendance Analysis

```python

low_attendance_students = df[df['attendance_percentage'] < 75]

if not low_attendance_students.empty:
    print("Students with attendance below 75%:")
    display(low_attendance_students[['name', 'department', 'attendance_percentage']])
else:
    print("All students have an attendance percentage above 75%.")
```
